```python
import pandas as pd
import os
```

## Load Data
There are two kinds of dataset, one is sone_data, another one is log_data, but both they are in json file format

```python
cwd = os.getcwd()
song_path = cwd + '/data/song_data/A'
song_path_list = [['A',['A','B','C']],['B',['A','B','C']]]
log_path = cwd + '/data/log_data/2018/11'
```

```python
song_df = pd.DataFrame()
for second_path in song_path_list:
    for third_path in second_path[1]:
        tem_path = song_path + '/' + second_path[0] + '/' + third_path
        tem_files = os.listdir(tem_path)
        for tem_file in tem_files :
            if tem_file[-5:] == '.json' :
                tem_df = pd.read_json(tem_path+'/'+tem_file, lines=True)
                song_df = pd.concat([song_df,tem_df],axis=0,ignore_index=True)
```

```python
log_df = pd.DataFrame()
tem_files = os.listdir(log_path)
for tem_file in tem_files :
    if tem_file[-5:] == '.json' :
        tem_df = pd.read_json(log_path+'/'+tem_file, lines=True)
        log_df = pd.concat([log_df,tem_df],axis=0,ignore_index=True)
```
