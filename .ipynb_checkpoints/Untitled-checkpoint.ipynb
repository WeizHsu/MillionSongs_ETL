```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()  # Get the current working directory (cwd)
files = os.listdir(cwd+'/data')
```

```python
files.remove('address-1.csv')
address = pd.read_csv('data/'+'address-1.csv')
for file in files:
    tem = pd.read_csv('data/'+file)
    address = pd.concat([address,tem],axis=0,ignore_index=True)
address = address.drop_duplicates().reset_index(drop=True)
```

```python
address_id = [i for i in range(1,13001)]
```

```python
postcode = address[['Postcode','state']].drop_duplicates().reset_index(drop=True)
postcode = postcode.rename(columns = {'state':'State_Code'})
```

```python
postcode.to_csv(cwd+'/postcode.csv',index=False)
```

```python
state_code = list(set(list(postcode.State_Code)))
state_name = state_code.copy()
```

```python
state = pd.DataFrame()
state['State_Code'] = state_code
state['State_Name'] = state_name
state.State_Name.replace({'NSW':'New South Wales', 'VIC':'Victoria', 'WA':'Western Australia', 'SA':'South Australia', 'TAS':'Tasmania', 'QLD':'Queensland'},inplace=True)
state
```

```python
state.to_csv(cwd+'/state.csv',index=False)
```
